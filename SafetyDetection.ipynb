```python
import pandas as pd
#data preprocessing
df= pd.read_csv("crime_dataset_india1.csv")
df
```

```python
#df.head()
#df['Crime Description'].unique()
```

```python
df=df.dropna(subset=['City','Time of Occurrence','Victim Age','Victim Gender','Crime Domain'])
#droping null values
```

```python
df=df[['City','Time of Occurrence','Victim Age','Victim Gender','Crime Domain']]
#jo columns chaiye vo select krliye
```

```python
df
```

```python
#label encoding (convert text to numbers)
from sklearn.preprocessing import LabelEncoder
le_city=LabelEncoder()
df['City']=le_city.fit_transform(df['City'])
le_gender=LabelEncoder()
df['Victim Gender']=le_gender.fit_transform(df['Victim Gender'])
# date time column fix kia , date aur time dono ko alg alg kiya aur date time column full ko drop kr diya
df['Time of Occurrence']=pd.to_datetime(df['Time of Occurrence'],errors='coerce')
df['Time']=df['Time of Occurrence'].dt.hour
df=df.drop(columns=['Time of Occurrence'])
print(le_city.classes_)
```

```python
#print(df.isnull().sum()) 
#hmare paass time me 23k rows NaN thi to hmne usme most common time daal diya from our dataset jiske liye hmne uska mode nikala
mct=df['Time'].mode()[0]
print("mct",mct)
#hmara mct aya 3.0 jiska mtlb 3 bje
#now we will replace NaN values with mct
df["Time"].fillna(mct)
```

```python
print(df['Time'].isnull().sum())
df
```

```python
# target column , voilent crime ko unsafe(0) and other crime ko safe(1)
df['is_safe']=df['Crime Domain'].apply(lambda x:0 if str(x).strip().lower() =='violent crime' else 1)
```

```python
# unsafe_keywords = ['KIDNAPPING', 'BURGLARY','ASSAULT', 'VEHICLE - STOLEN','EXTORTION', 'PUBLIC INTOXICATION', 'SEXUAL ASSAULT',
#        'DRUG OFFENSE','ILLEGAL POSSESSION']

# def mark_safe(text):
#     text = str(text).lower()
#     return 0 if any(word in text for word in unsafe_keywords) else 1

# df['is_safe'] = df['Crime Description'].apply(mark_safe)
print(df['is_safe'].value_counts())
```

```python
# train model (ogistic regression use kiya ha becz hm ak percentage de rahe h as a confidence score for safe and unself thing)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x=df[['City','Time','Victim Age','Victim Gender']]
y=df['is_safe']
#clearing nan values 
data=pd.concat([x,y],axis=1)
data=data.dropna()
x_clean=data[x.columns]
y_clean=data[y.name]
#split data for testing and training 
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)

#train model
# model=LogisticRegression(class_weight='balanced',max_iter=1000)
# model.fit(x_clean , y_clean)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
print(y.value_counts())      # distribution before the split
print(y_train.value_counts())  # distribution in the training fold
```

```python
def predict_safety(city,time_hour,age,gender):
    try:
        if city not in le_city.classes_:
            return{"OOPS! City not yet found . We're reaching there soon !"}
        if gender not in le_gender.classes_:
            return("Wrong choice entered")
        city_encoded=le_city.transform([city])[0]
        gender_encoded=le_gender.transform([gender])[0]
        input_data=pd.DataFrame([{
            "City":city_encoded,
            "Time":time_hour,
            "Victim Age":age,
            "Victim Gender":gender_encoded
        }])
        prob=model.predict_proba(input_data)[0]
        return {
           "safe ":float(round(prob[1]*100,2)),
           "unsafe":float(round(prob[0]*100,2)),
           "interpretation":"Likely Safe" if prob[1]>0.5 else "Likely unsafe, try to avoid!"
       }

    except Exception as e:
         return {"error":str(e)}
```

```python

# result=predict_safety("Mumbai",18,15,"Female")
for i in range(len(le_city.classes_)):
    print(i,le_city.inverse_transform([i])[0])

print(le_gender.classes_)
```

```python
result=predict_safety("Varanasi",13,45,'F')
print(result)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_clean)
accuracy = accuracy_score(y_clean, y_pred)
print("Model Accuracy:", round(accuracy * 100, 2), "%")
```
